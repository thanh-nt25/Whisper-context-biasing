```python
from datasets import load_dataset

dataset = load_dataset("Jzuluaga/uwb_atcc")
dataset
```

```python
train = dataset["train"]
test = dataset["test"]
train[0]
```

```python
for i in range(11291):
  print(train[i]["text"])
```

```python
def compute_total_duration_in_hours(split):
    total_samples = 0
    sampling_rate = None

    for item in split:
        audio = item["audio"]
        total_samples += len(audio["array"])
        sampling_rate = audio["sampling_rate"]

    total_seconds = total_samples / sampling_rate
    total_hours = total_seconds / 3600
    return total_hours


train_hours = compute_total_duration_in_hours(train)
test_hours = compute_total_duration_in_hours(test)

print(f"Tng thi lng audio (train): {train_hours:.2f} gi")
print(f"Tng thi lng audio (test): {test_hours:.2f} gi")
```
