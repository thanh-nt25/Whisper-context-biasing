```python
import os
import json


root_dir = "medical-united-syn-med-75/train"
output_file = "all_texts_train.jsonl"


with open(output_file, "w", encoding="utf-8") as out_file:
    index = 0

    
    for dirpath, dirnames, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.endswith(".json"):
                filepath = os.path.join(dirpath, filename)
                try:
                    with open(filepath, "r", encoding="utf-8") as f:
                        data = json.load(f)
                        if "text" in data:
                            json_line = {
                                "id": f"{index:06d}",
                                "file": filename,
                                "text": data["text"].strip()
                            }
                            out_file.write(json.dumps(json_line, ensure_ascii=False) + "\n")
                            index += 1
                except Exception as e:
                    print(f"Lỗi khi đọc file {filepath}: {e}")

print(f"Đã ghi {index} dòng JSON vào '{output_file}'")
```

```python
import json
import random


train_path = "all_texts_train.jsonl"
dev_path = "all_texts_dev.jsonl"
output_path = "train_dev_5000_suffer.jsonl"


train_sample_size = 4250
dev_sample_size = 750


with open(train_path, "r", encoding="utf-8") as f:
    train_data = [json.loads(line.strip()) for line in f]


with open(dev_path, "r", encoding="utf-8") as f:
    dev_data = [json.loads(line.strip()) for line in f]


train_sample = random.sample(train_data, train_sample_size)
dev_sample = random.sample(dev_data, dev_sample_size)


combined_data = train_sample + dev_sample
random.shuffle(combined_data)


with open(output_path, "w", encoding="utf-8") as out_f:
    for idx, item in enumerate(combined_data, 1):
        out_f.write(json.dumps(item, ensure_ascii=False) + "\n")
        if idx % 100 == 0 or idx == 1 or idx == len(combined_data):
            print(f"✅ Đã ghi {idx}/{len(combined_data)} dòng")

print(f"\nHoàn tất! Đã ghi tổng cộng {len(combined_data)} dòng vào {output_path}")
```

```python
import re

def get_next_id(output_path):
    if not os.path.exists(output_path):
        return "000001"
    
    last_id = "000000"
    with open(output_path, "r", encoding="utf-8") as f:
        for line in f:
            try:
                item = json.loads(line)
                current_id = item["id"]
                if current_id > last_id:
                    last_id = current_id
            except:
                continue

    
    try:
        next_id = str(int(last_id) + 1).zfill(len(last_id))
        return next_id
    except:
        return "000001"
```

```python
import os
import json
from openai import OpenAI

client = OpenAI(api_key=os.getenv("GPT_API_TOKEN"))

input_path = "all_texts_train.jsonl"
output_path = "all_texts_train_description.jsonl"
start_from_id = get_next_id(output_path)
end_at_id = "034357"
BATCH_SIZE = 20

print(f"Continuing from ID: {start_from_id} → End at ID: {end_at_id}")

# Tập hợp các ID đã xử lý
existing_ids = set()
if os.path.exists(output_path):
    with open(output_path, "r", encoding="utf-8") as f:
        for line in f:
            try:
                item = json.loads(line)
                existing_ids.add(item["id"])
            except:
                continue

def build_batch_prompt(texts):
    prompt = """You are given several medical transcripts. For each one, return a numbered, concise yet medically-informative description that:

- Identifies relevant terms from these categories: drugchemical, diagnostics, meddevicetechnique
- Explains the medical role and purpose of the term(s)
- Highlights useful information for speech recognition (e.g. brand vs. generic name, specific context, affected body parts)
- If no specialized terms exist, briefly describe the medical context

Example:
1. "Domperidone SR is prescribed for treating gastrointestinal issues like nausea and vomiting."
→ Domperidone SR, a prokinetic drug, is used to alleviate gastrointestinal symptoms such as nausea and vomiting, aiding in the management of digestive disorders.

Now process the following transcripts:
"""
    for i, text in enumerate(texts, 1):
        prompt += f"{i}. {text}\n"

    prompt += "\nReturn your output in the format:\n1. ...\n2. ...\n..."
    return prompt

def process_batch(entries_batch, outfile):
    texts = [e["text"] for e in entries_batch]
    prompt = build_batch_prompt(texts)

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.5,
    )

    response_lines = response.choices[0].message.content.strip().split("\n")
    descriptions = {}
    for line in response_lines:
        if "." in line:
            idx_str, desc = line.split(".", 1)
            try:
                idx = int(idx_str.strip())
                descriptions[idx] = desc.strip()
            except:
                continue

    for i, entry in enumerate(entries_batch):
        desc = descriptions.get(i + 1)
        if desc:
            entry["description"] = desc
            outfile.write(json.dumps(entry, ensure_ascii=False) + "\n")
            print(f"✅ Done ID: {entry['id']}")
        else:
            print(f"⚠️  Missing description for ID: {entry['id']}")

# ==================== MAIN LOOP ====================
entries_batch = []

with open(input_path, "r", encoding="utf-8") as infile, open(output_path, "a", encoding="utf-8") as outfile:
    for line in infile:
        try:
            entry = json.loads(line)
            entry_id = entry["id"]

            if entry_id < start_from_id or entry_id in existing_ids:
                continue

            entries_batch.append(entry)

            if len(entries_batch) == BATCH_SIZE:
                process_batch(entries_batch, outfile)
                entries_batch = []

            if end_at_id and entry_id >= end_at_id:
                if entries_batch:
                    process_batch(entries_batch, outfile)
                print(f"🛑 Reached end_at_id: {end_at_id}. Stopping.")
                break

        except Exception as e:
            print(f"❌ Error at ID {entry.get('id', 'unknown')}: {e}")
```

```python
import os
import json
import re
from openai import OpenAI

# ========== CONFIG ==========
client = OpenAI(api_key=os.getenv("GPT_API_TOKEN"))
input_path = "all_texts_test.jsonl"
output_path = "all_tests_with_bias_list.jsonl"
start_from_id = "000000"
end_at_id = "005114"
BATCH_SIZE = 20
# ============================

print(f"🔄 Processing from ID: {start_from_id} → End at ID: {end_at_id}")

# Ghi nhớ các ID đã được xử lý
existing_ids = set()
if os.path.exists(output_path):
    with open(output_path, "r", encoding="utf-8") as f:
        for line in f:
            try:
                item = json.loads(line)
                existing_ids.add(item["id"])
            except:
                continue

# ===== Prompt Builder =====
def build_batch_prompt(texts):
    prompt = """You are a medical labeling assistant for speech recognition evaluation.

Your task is to extract **bias words** from each medical transcript. Focus ONLY on specialized medical terms from these 3 categories:

- DRUGCHEMICAL (e.g., Amoxicillin, Azithromycin, Metformin)
- DIAGNOSTICS (e.g., MRI, Endoscopy, Tomography, Angiography)
- MEDDEVICETECHNIQUE (e.g., Catheter, Defibrillator, Ventilator, Endotracheal)

DO NOT include general or common words that Whisper would already recognize (e.g., fever, doctor, treatment, medicine).

Only return real, medically specific terms. If none exist, return an empty list.

Now extract bias words from the following transcripts:
"""
    for i, text in enumerate(texts, 1):
        prompt += f"{i}. {text}\n"

    prompt += "\nReturn only JSON lists like:\n1. [\"word1\", \"word2\"]\n2. [\"word3\"]\n..."
    return prompt

# ===== GPT Batch Handler =====
def process_batch(entries_batch, outfile):
    texts = [e["text"] for e in entries_batch]
    prompt = build_batch_prompt(texts)

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": prompt}],
            temperature=0,
        )
        content = response.choices[0].message.content.strip()
    except Exception as e:
        print(f"❌ GPT API Error: {e}")
        return

    bias_dict = {}
    for line in content.splitlines():
        match = re.match(r"(\d+)\.\s*(\[[^\]]*\])", line)
        if match:
            idx = int(match.group(1))
            try:
                bias_list = json.loads(match.group(2))
                bias_dict[idx] = bias_list
            except:
                continue

    for i, entry in enumerate(entries_batch):
        bias_words = bias_dict.get(i + 1, [])
        result = {
            "id": entry["id"],
            "text": entry["text"],
            "bias_words": bias_words
        }
        outfile.write(json.dumps(result, ensure_ascii=False) + "\n")
        print(f"✅ ID {entry['id']} → {bias_words}")

# ===== MAIN LOOP =====
entries_batch = []
with open(input_path, "r", encoding="utf-8") as infile, open(output_path, "a", encoding="utf-8") as outfile:
    for line in infile:
        try:
            entry = json.loads(line)
            entry_id = entry["id"]

            if entry_id < start_from_id or entry_id in existing_ids:
                continue

            entries_batch.append(entry)

            if len(entries_batch) == BATCH_SIZE:
                process_batch(entries_batch, outfile)
                entries_batch = []

            if end_at_id and entry_id >= end_at_id:
                if entries_batch:
                    process_batch(entries_batch, outfile)
                print(f"🛑 Reached end_at_id: {end_at_id}. Done.")
                break

        except Exception as e:
            print(f"❌ Error at ID {entry.get('id', 'unknown')}: {e}")
```
